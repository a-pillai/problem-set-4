```python
import urllib.request, urllib.parse, urllib.error
import json
import matplotlib.pyplot as plt
```

```python
serviceurl = 'http://makeup-api.herokuapp.com/api/v1/products.json'
while True:
    brand = input('Enter a brand name: ')
    if len(brand) < 1: break

    url = serviceurl + "?" + urllib.parse.urlencode(
        {'brand': brand})
    print('Retrieving', url)
    uh = urllib.request.urlopen(url)
    data = uh.read().decode()
    print('Retrieved', len(data), 'characters')
    try:
        js = json.loads(data)
    except:
        js = None
    name=[]
    price=[]
    for product in js:
        print(product['name'], ',price:', product['price'])# retrieved the produce name and price in brande maybelline
        name.append(product['name'])
        price.append(float(product['price']))
    plt.figure(figsize=(13,4)) # plot graph of name and price
    plt.plot(name,price)
    plt.xticks(rotation=90)
    plt.xticks(fontsize=10)
    plt.show()
```
